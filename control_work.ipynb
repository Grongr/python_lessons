# Контрольная работа!

### Задача 1.
Напишите программу, которая принимает два числа a и b и вычисляет:

Если a > b, результат выражения $(a^2 - b^3) / 2$

Если a <= b, результат выражения $(a^2 + b^3) / 3$

---

Пример ввода:
```text
a = 5
b = 3
```

Пример вывода:
```text
-1
```

---

### Задача 2.

Напишите программу, которая для числа, полученного из стандартного ввода, определяет:

1. Если оно положительное, является ли оно четным.
2. Если оно отрицательное, делится ли оно на 11, и если да, то с каким остатком.

---

Пример ввода:
```text
 2
```

Пример вывода:
```text
Четное
```

---------------------------------

Пример ввода:
```text
-11
```

Пример вывода:
```text
Отрицательное, делится на 11.
```

### Задача 5.

Напишите программу, которая:

1. Создаёт список из первых N натуральных чисел.
2. Формирует новый список, где остаются числа, дающие остаток 1 при делении на 3.
3. Вычисляет сумму квадратов элементов нового списка.

Пример ввода:
```text
N = 6
```

Пример вывода:
```text
Исходный список: [1, 2, 3, 4, 5, 6]
Числа с остатком 1: [1, 4]
Сумма квадратов: 17
```
